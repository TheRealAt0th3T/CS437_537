Given a query q, create a set of documents CR comprised of all the documents in DC that contain each of the terms in q. If there are less than 50 documents, then you will need to consider resources in DC that contain n-1 terms in q (all the combinations).

```python
import pandas as pd
import re, string
import math
import time
import json

start = time.time()
with open('Computed/wiki_dict.json') as json_file:
    wiki_dict = json.load(json_file)
wiki = pd.read_csv('Computed/wiki.csv')[['content','title','id','max_occur_words','max_occur_number']]
print(wiki)
end = time.time()
print(end - start)
```

```python
def minusOne(array):
    ret = []
    for a in array:
        ret.append(int(a)-1)
    return ret

# Takes a word, returns all documents as dataframe that contains the word (Wiki)
def containsIDWiki(qArray):
    keys = []
    for qWord in qArray.split(" "):
        print(qWord)
        keys += minusOne(list(wiki_dict[str(qWord)].keys()))
#     print("before: " + str(keys))
    keys = list(dict.fromkeys(keys)) # remove duplicates
    keys.sort()
#     print("after: " + str(keys))

    return wiki.iloc[keys]
```

## Freq

freq(w, d) is the number of times **w** appears in **d**

In our case, w is a word in the query, and d is the CR that we had. 

```python
def getFreq(word,doc_id):
    if str(doc_id) in wiki_dict[word].keys():
        return wiki_dict[word][str(doc_id)]
    else:
        return 0

print(getFreq('morocco', 1))
```

## Maxd

Maxd is the number of times the most frequently-occurred term appears in d (which is constant for each document)

```python
def getMaxd(doc_id):
    return CR['max_occur_number'].iloc[doc_id-1]

print(getMaxd(1))
```

## N

N is the number of documents in DC

```python
def getN():
    return len(wiki)

print(getN())
```

## nw

nw is the number of documents in DC in which w appears at least once

```python
def getnw(word):
    return len(list(wiki_dict[word].keys()))

print(getnw('morocco'))
```

## TF-IDF

Application for the whole query

```python
def getTFIDF(word, doc_id):
    TF = getFreq(word, doc_id)/getMaxd(doc_id)
    IDF = math.log2(getN()/getnw(word))
    return TF*IDF

word = "morocco"
doc_id = 1
print(getTFIDF(word,doc_id))
```
